```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
from google.colab import files
import pandas as pd

# Upload file
uploaded = files.upload()

# Load dataset into DataFrame
df = pd.read_csv("admission_enquiries_btech.csv")  # Replace with the actual filename
```

# 1. Basic Data Overview

```python
# 1. Basic Data Overview
print("Dataset Information:")
df.info()
print("\nSummary Statistics:")
print(df.describe())
```

# 2. Check for Missing Values


```python
# 2. Check for Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

# 3. Department-wise Analysis

```python
# 3. Department-wise Analysis
plt.figure(figsize=(8, 4))
sns.countplot(x=df['Department'], palette='viridis')
plt.title("Number of Students per Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# 4. Interest Area Analysis

```python
# 4. Interest Area Analysis
plt.figure(figsize=(10, 4))
sns.countplot(x=df['Interest Area'], palette='coolwarm', order=df['Interest Area'].value_counts().index)
plt.title("Interest Area Distribution")
plt.xlabel("Interest Area")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# 5. CGPA vs B.Tech Score Relationship

```python

# 5. CGPA vs B.Tech Score Relationship
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df['CGPA'], y=df['B.Tech Score'], alpha=0.5)
plt.title("CGPA vs. B.Tech Score")
plt.xlabel("CGPA")
plt.ylabel("B.Tech Score")
plt.show()
```

# 6. Top Students Based on B.Tech Score

```python
# 6. Top Students Based on B.Tech Score
top_students = df.nlargest(10, 'B.Tech Score')
print("\nTop 10 Students Based on B.Tech Score:")
print(top_students[['Name', 'Department', 'CGPA', 'B.Tech Score', 'Interest Area']])
```

# 7. Boxplots for Distribution Analysis

```python
# 7. Boxplots for Distribution Analysis
plt.figure(figsize=(10, 4))
sns.boxplot(data=df[['CGPA', 'B.Tech Score']], palette='Set2')
plt.title("Distribution of CGPA and B.Tech Score")
plt.show()
```

# 8. Heatmap for Correlations

```python
# 8. Heatmap for Correlations
plt.figure(figsize=(6, 4))
sns.heatmap(df[['CGPA', 'B.Tech Score']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
